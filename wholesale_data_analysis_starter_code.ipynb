## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df[['unit_price','qty','subtotal']].head()
```

```python
# Create a column for shipping price. Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def final_shipping_price(x):
    if x > 50:
        return 7
    else:
        return 10

df['total_weight'] = df['qty'] * df['unit_weight']
df['shipping_price'] = df['total_weight'] * df['total_weight'].apply(final_shipping_price)
df[['unit_price','unit_weight','qty','total_weight','shipping_price']].head()


```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['total_price'] = df['subtotal'] + df['shipping_price'] + (df['subtotal'] + df['shipping_price']) * 0.0925
df[['subtotal','shipping_price','total_price']].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df[['unit_cost','qty','shipping_price','line_cost']].head()
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df['total_price'] - df['line_cost']
df[['line_cost','total_price','line_profit']].head()

```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
top_5_df = df.groupby('client_id').agg({'total_price':'sum'}).reset_index()
top_5_df[top_5_df['client_id'].isin(top_5)]

```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

top_5_summary = df.groupby('client_id').agg({'qty':'sum','shipping_price':'sum','total_price':'sum','line_profit':'sum'}).reset_index()
top_5_summary[top_5_summary['client_id'].isin(top_5)]
```

```python
# Format the data and rename the columns to names suitable for presentation.
top_5_summary.columns = ['Client ID','Total Units Purchased','Total Shipping Price','Total Revenue','Total Profit'] 
top_5_summary['Total Units Purchased'] = top_5_summary['Total Units Purchased'].apply(lambda x: "{x:,.0f}")
top_5_summary['Total Shipping Price'] = top_5_summary['Total Shipping Price'].apply(lambda x: "${x:,.2f}")
top_5_summary['Total Revenue'] = top_5_summary['Total Revenue'].apply(lambda x: "${x:,.2f}")
top_5_summary['Total Profit'] = top_5_summary['Total Profit'].apply(lambda x: "${x:,.2f}")
top_5_summary

# Define the money columns. 
money_columns = ['unit_price','unit_cost','subtotal','shipping_price','total_price','line_cost','line_profit']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(x):
    return "${x/1000000:,.2f}M"

# Apply the currency_format_millions function to only the money columns. 
df[money_columns] = df[money_columns].applymap(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
df.columns = ['Order ID', 'Client ID', 'Category', 'Subcategory', 'Product', 'Unit Price', 'Unit Cost', 'Unit Weight', 'Qty', 'Subtotal', 'Total Weight', 'Shipping Price', 'Total Price', 'Line Cost', 'Line Profit']

```
